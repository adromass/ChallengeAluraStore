```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
import numpy as np
from matplotlib.ticker import MaxNLocator, FuncFormatter

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# 1. Calcular el ingreso total por tienda (suma de la columna 'Precio')
ingresoTotalTienda = tienda['Precio'].sum()
ingresoTotalTienda2 = tienda2['Precio'].sum()
ingresoTotalTienda3 = tienda3['Precio'].sum()
ingresoTotalTienda4 = tienda4['Precio'].sum()

# 2 Función para contar productos por categoría
def ventas_por_categoria(df):
    conteo = df.groupby('Categoría del Producto')['Producto'].count().reset_index()
    conteo.columns = ['Categoría', 'Ventas']
    conteo = conteo.sort_values('Ventas', ascending=False).reset_index(drop=True)
    conteo.index = conteo.index + 1
    display(Markdown("### Ventas por categoría"))
    display(conteo)
    return

# 3 Función para calcular promedio de calificaciones por tienda
def promedio_calificaciones(df):
    promedio = df['Calificación'].mean()
    return f"{promedio:,.2f}"

# 4 Función para analizar productos más/menos vendidos
def analizar_ventas_productos(df):
    # Contar ventas por producto
    ventas_por_producto = df['Producto'].value_counts().reset_index()
    ventas_por_producto.columns = ['Producto', 'Ventas']
    
    # Ordenar de mayor a menor
    ventas_por_producto = ventas_por_producto.sort_values('Ventas', ascending=False)
    
    # Obtener top 5 más vendidos y top 5 menos vendidos
    top_mas_vendidos = ventas_por_producto.head(10)
    top_menos_vendidos = ventas_por_producto.tail(10).sort_values('Ventas', ascending=True).reset_index(drop=True)
    
    top_mas_vendidos.index = top_mas_vendidos.index + 1
    top_menos_vendidos.index = top_menos_vendidos.index + 1
    
    display(Markdown("### Productos Más Vendidos"))
    display(top_mas_vendidos)
    display(Markdown("### Productos Menos Vendidos"))
    display(top_menos_vendidos)
    
    return

# 5 Función para calcular costo de envío promedio
def costo_envio_promedio(df):
    promedio = df['Costo de envío'].mean()
    return promedio
    #f"${promedio:,.2f}"
    #print(f"Costo de envío promedio en {nombre_tienda}: ${promedio:.2f}")

#funcion para presentar datos como tabla jupyter
def crear_tabla (nombres, *datos):
    df=pd.DataFrame(datos, columns=nombres)
    #display(df)
    display(df.style.hide(axis="index").hide(axis="columns"))
    return

#funcion para crear graficas
def analisis_graficas(tiendas, ingresos, costo_envio, ventas_trimestrales):

    #grafica de barras para los ingresos totales por tienda
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.bar(
        tiendas, 
        [v / 1e6 for v in ingresos], 
        color = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
    )
    ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.0f}'))
    plt.ylabel("Monto $ (en millones)")
    plt.title('Ingresos Totales')
    plt.show()

    #grafica de barras para los costos de envio promedio por tienda
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.bar(tiendas, costo_envio, color = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
    ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.0f}'))
    plt.ylabel("Monto $")
    plt.title('Costos de Envio Promedio')
    plt.show()

    #grafica circular para la utlidad neta por tienda
    fig, ax = plt.subplots(figsize=(4, 4))
    utilidad = [d1 + d2 for d1, d2 in zip(ingresos, costo_envio)]
    
    plt.pie(utilidad, labels=tiendas, autopct='%1.1f%%', startangle=90)
    plt.title("Utilidad por Tienda")
    plt.axis('equal')  # Para que el gráfico se vea circular
    plt.show()

    #grafica de dispersion    
    # Tipos dentro de cada lista

    fig, ax = plt.subplots(figsize=(9, 6))
    offset=0.5

    for i, serie in enumerate(ventas_trimestrales):
        trimestres = serie.index.astype(str)
        ventas = serie.values
        x = np.arange(len(ventas))

        # Línea de ventas (usando x numérico para graficar)
        ax.plot(
            x,
            ventas,
            label=tiendas[i],
            alpha=0.5
        )

        # Línea de regresión lineal (tendencia)
        coef = np.polyfit(x, ventas, 1)
        tendencia = np.polyval(coef, x)

        ax.plot(
            x,
            tendencia + offset * i,
            linestyle='--',
            color=ax.lines[-1].get_color(),
            label=f"Tendencia {tiendas[i]}"
        )
    
    # Poner etiquetas legibles en eje X con los trimestres
    ax.set_xticks(np.arange(len(trimestres)))
    ax.set_xticklabels(trimestres, rotation=45)
    
    ax.set_title("Ventas Trimestrales por Tienda")
    ax.set_xlabel("Trimestre")
    ax.set_ylabel("Cantidad de Ventas")
    ax.legend()
    ax.grid(True)
    plt.tight_layout()
    plt.show()

#Analisis tienda 1

display(Markdown("# Análisis De Tienda 1"))

crear_tabla(
    ["nombre", "datos"],
    ["Ingresos Totales", f"${ingresoTotalTienda:,.2f}"],
    ["Costo de envio promedio", f"${costo_envio_promedio(tienda):,.2f}"],
    ["Promedio de calificación", promedio_calificaciones(tienda)]
)

ventas_por_categoria(tienda)

analizar_ventas_productos(tienda)

#Analisis tienda 2

display(Markdown("# Análisis De Tienda 2"))

crear_tabla(
    ["nombre", "datos"],
    ["Ingresos Totales", f"${ingresoTotalTienda2:,.2f}"],
    ["Costo de envio promedio", f"${costo_envio_promedio(tienda2):,.2f}"],
    ["Promedio de calificación", promedio_calificaciones(tienda2)]
)

ventas_por_categoria(tienda2)

analizar_ventas_productos(tienda2)

#Analisis tienda 3

display(Markdown("# Análisis De Tienda 3"))

crear_tabla(
    ["nombre", "datos"],
    ["Ingresos Totales", f"${ingresoTotalTienda3:,.2f}"],
    ["Costo de envio promedio", f"${costo_envio_promedio(tienda3):,.2f}"],
    ["Promedio de calificación", promedio_calificaciones(tienda3)]
)

ventas_por_categoria(tienda3)

analizar_ventas_productos(tienda3)

#Analisis tienda 4

display(Markdown("# Análisis De Tienda 4"))

crear_tabla(
    ["nombre", "datos"],
    ["Ingresos Totales", f"${ingresoTotalTienda4:,.2f}"],
    ["Costo de envio promedio", f"${costo_envio_promedio(tienda4):,.2f}"],
    ["Promedio de calificación", promedio_calificaciones(tienda4)]
)

ventas_por_categoria(tienda4)

analizar_ventas_productos(tienda4)


#Graficas para analizar las tiendas

display(Markdown("# Análisis De Tiendas"))

analisis_graficas(
    ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda4'],
    [ingresoTotalTienda, ingresoTotalTienda2, ingresoTotalTienda3, ingresoTotalTienda4],
    [costo_envio_promedio(tienda), costo_envio_promedio(tienda2), 
     costo_envio_promedio(tienda3), costo_envio_promedio(tienda4)],
    [pd.to_datetime(tienda['Fecha de Compra'], dayfirst=True).dt.to_period('Q').value_counts().sort_index(), 
     pd.to_datetime(tienda2['Fecha de Compra'], dayfirst=True).dt.to_period('Q').value_counts().sort_index(), 
     pd.to_datetime(tienda3['Fecha de Compra'], dayfirst=True).dt.to_period('Q').value_counts().sort_index(), 
     pd.to_datetime(tienda4['Fecha de Compra'], dayfirst=True).dt.to_period('Q').value_counts().sort_index()]
    #pd.to_datetime:convierte str de 'Fecha de Compra' a formato fecha
    #dayfirst=true le indica a pandas que el primer numero es el dia.
    #pd.to_period('Q') convierte la fecha a trimestre
    #.value_counts() cuenta cuantas veces se repiten trimestres y .sort_index() los ordena
)
```
