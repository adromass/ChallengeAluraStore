```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()

# 1. Calcular el ingreso total por tienda (suma de la columna 'Precio')
ingresoTotalTienda = tienda['Precio'].sum()
ingresoTotalTienda2 = tienda2['Precio'].sum()
ingresoTotalTienda3 = tienda3['Precio'].sum()
ingresoTotalTienda4 = tienda4['Precio'].sum()

print(ingresoTotalTienda)
print(ingresoTotalTienda2)
print(ingresoTotalTienda3)
print(ingresoTotalTienda4)

# Funci√≥n para contar productos por categor√≠a
def ventas_por_categoria(df, nombre_tienda):
    conteo = df.groupby('Categor√≠a del Producto')['Producto'].count().reset_index()
    conteo.columns = ['Categor√≠a', 'Ventas']
    print(f"\nVentas por categor√≠a en {nombre_tienda}:")
    print(conteo.sort_values('Ventas', ascending=False))  # Ordenar de mayor a menor

# Aplicar la funci√≥n a cada tienda
ventas_por_categoria(tienda, "Tienda 1")
ventas_por_categoria(tienda2, "Tienda 2")
ventas_por_categoria(tienda3, "Tienda 3")
ventas_por_categoria(tienda4, "Tienda 4")

# Funci√≥n para calcular promedio de calificaciones por tienda
def promedio_calificaciones(df, nombre_tienda):
    if 'Calificaci√≥n' in df.columns:  # Ajusta el nombre seg√∫n tus datos
        promedio = df['Calificaci√≥n'].mean()
        print(f"\nPromedio de calificaciones en {nombre_tienda}: {promedio:.2f}")
    else:
        print(f"\nNo se encontr√≥ columna 'Calificaci√≥n' en {nombre_tienda}")

# Aplicar a cada tienda
promedio_calificaciones(tienda, "Tienda 1")
promedio_calificaciones(tienda2, "Tienda 2")
promedio_calificaciones(tienda3, "Tienda 3")
promedio_calificaciones(tienda4, "Tienda 4")

# Funci√≥n para analizar productos m√°s/menos vendidos
def analizar_ventas_productos(df, nombre_tienda):
    # Verificar si la columna 'Producto' existe
    if 'Producto' not in df.columns:
        print(f"\nError en {nombre_tienda}: No existe columna 'Producto'")
        return None, None
    
    # Contar ventas por producto
    ventas_por_producto = df['Producto'].value_counts().reset_index()
    ventas_por_producto.columns = ['Producto', 'Total_Ventas']
    
    # Ordenar de mayor a menor
    ventas_por_producto = ventas_por_producto.sort_values('Total_Ventas', ascending=False)
    
    # Obtener top 5 m√°s vendidos y top 5 menos vendidos
    top_mas_vendidos = ventas_por_producto.head(5)
    top_menos_vendidos = ventas_por_producto.tail(5)
    
    return top_mas_vendidos, top_menos_vendidos

# Funci√≥n para graficar resultados
def graficar_ventas(top_mas, top_menos, nombre_tienda):
    plt.figure(figsize=(12, 6))
    
    # Gr√°fico de barras para m√°s vendidos (verde)
    plt.bar(top_mas['Producto'], top_mas['Total_Ventas'], color='green', alpha=0.6, label='M√°s vendidos')
    
    # Gr√°fico de barras para menos vendidos (rojo)
    plt.bar(top_menos['Producto'], top_menos['Total_Ventas'], color='red', alpha=0.6, label='Menos vendidos')
    
    plt.title(f'Productos m√°s y menos vendidos - {nombre_tienda}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad vendida')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Analizar cada tienda (usando las variables originales)
for nombre, df in [('Tienda 1', tienda), 
                   ('Tienda 2', tienda2), 
                   ('Tienda 3', tienda3), 
                   ('Tienda 4', tienda4)]:
    
    print(f"\n{'='*50}\nAn√°lisis de: {nombre}\n{'='*50}")
    
    mas_vendidos, menos_vendidos = analizar_ventas_productos(df, nombre)
    
    if mas_vendidos is not None:
        print("\nüî• TOP 5 PRODUCTOS M√ÅS VENDIDOS:")
        print(mas_vendidos)
        
        print("\n‚ùå TOP 5 PRODUCTOS MENOS VENDIDOS:")
        print(menos_vendidos)
        
        graficar_ventas(mas_vendidos, menos_vendidos, nombre)
    
    # Funci√≥n para calcular costo de env√≠o promedio
def costo_envio_promedio(df, nombre_tienda):
    if 'Costo de env√≠o' in df.columns:  # Verificar si existe la columna
        promedio = df['Costo de env√≠o'].mean()
        print(f"Costo de env√≠o promedio en {nombre_tienda}: ${promedio:.2f}")
    else:
        print(f"‚ö†Ô∏è {nombre_tienda} no tiene columna 'Costo de env√≠o'")

# Analizar cada tienda
print("=== AN√ÅLISIS DE COSTOS DE ENV√çO ===")
costo_envio_promedio(tienda, "Tienda 1")
costo_envio_promedio(tienda2, "Tienda 2")
costo_envio_promedio(tienda3, "Tienda 3")
costo_envio_promedio(tienda4, "Tienda 4")
```
